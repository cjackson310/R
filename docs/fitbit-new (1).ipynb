```r
library(tidyverse)
library(scales)
library(lubridate)
```

```r
daily_activity <- read_csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv")
```

```r
daily_activity$Ymd <- mdy(daily_activity$ActivityDate)
```

```r
str(daily_activity)
```

```r
daily_activity %>% count(Ymd, sort = TRUE) %>%
ggplot() + geom_col(aes(x=Ymd, y=n)) + labs(title = "Count of Activities by Day")
```

```r
total_steps <- daily_activity %>% group_by(Ymd) %>%
summarize( Steps = sum(TotalSteps, na.rm=TRUE),
         Mean = mean(TotalSteps , na.rm = TRUE))
ggplot(total_steps) + geom_line(aes(x=Ymd,y=Steps)) +
labs(title = "Total Steps by Day") +
scale_y_continuous(labels = comma)
```

```r
ggplot(total_steps) + geom_line(aes(x=Ymd,y=Mean)) +
labs(title = "Average Steps by Day") +
scale_y_continuous(labels = comma)
```

```r
total_distance <- daily_activity %>% group_by(Ymd) %>%
summarize( Total = sum(TotalDistance, na.rm=TRUE),
         Mean = mean(TotalDistance , na.rm = TRUE))
```

```r
ggplot(total_distance) +
geom_line(aes(x=Ymd, y=Total)) +
labs(title = "Total Distance by Date", y= "Total Distance")
```

```r
ggplot(total_distance) +
geom_line(aes(x=Ymd, y=Mean)) +
labs(title = "Average Distance by Date", y= "Total Distance")
```
